```python
def predict(theta,X):
    return theta[0] + theta[1] * X
```

```python
def compute_error(hx,y):
    return hx - y
```

```python
def compute_cost(e,X):
    return (np.linalg.norm(e) ** 2) / (2 * X.shape[0])
```

```python
def compute_gradient_vector(e,X):
    d_theta_0 = np.sum(e) / X.shape[0]
    d_theta_1 = np.dot(e,X) / X.shape[0]
    gradient_vector = np.array([[d_theta_0], [d_theta_1]])
    return gradient_vector
```

```python
def compute_gradient_norm(gradient_vector):
    return np.linalg.norm(gradient_vector)
```

```python
def update(theta,learning_rate,gradient_vector):
    theta[0] -= learning_rate * gradient_vector[0]
    theta[1] -= learning_rate * gradient_vector[1]
    return theta
    
```

```python
def compute_gradient(X,y,learning_rate,max_iterations):
    loss =[] 
    theta = np.array([[0.0],[0.0]])
    all_theta_0 = []  
    all_theta_1 = []  
    all_hx = []  
    tolerance = 0.001
    first_iteration = True
    for iteration in range(max_iterations):
        print(f"******************* Iteration: {iteration} *******************************")
        
        hx = predict(theta,X)
        all_hx.append(hx)
        print(f"h(x):{hx} \n")
        
        e = compute_error(hx,y)
        print(f"Error vector: {e} \n")
        
        j = compute_cost(e,X)
        loss.append(j)
        print(f"j = {j} \n")
        
        gradient_vector = compute_gradient_vector(e,X)
        print(f"Gradient_vector:  {gradient_vector} \n")
        
        gradient_vector_norm = compute_gradient_norm(gradient_vector)
        print(f"Gradient_vector_norm: {gradient_vector_norm} \n")
        
        all_theta_0.append(theta[0][0])
        all_theta_1.append(theta[1][0])
        
        if not first_iteration:
            if abs(loss[iteration] - loss[iteration - 1]) < tolerance:
                print(f"Converged at iteration {iteration} with cost: {loss[iteration]} \n")
                break
            if gradient_vector_norm < tolerance:
                print(f"Stopping at iteration {iteration} due to small gradient norm: {gradient_vector_norm} \n")
                break
        
        theta = update(theta,learning_rate,gradient_vector)
        print(f"new theta_0 : {theta[0][0]}")
        print(f"new theta_1 : {theta[1][0]}")
        
        first_iteration = False
        
    print ("************************** Training Report ****************************")
    print(f"Gradient Descent converged after {len(all_theta_1)} epochs \n")
    print(f"Final theta_0: {theta[0][0]}")
    print(f"Final theta_1: {theta[1][0]} \n")

    return all_theta_0, all_theta_1, loss, all_hx, theta
        
```

#### Try your function with any input parameterss of your choice.
##### The following results uses alpha = 0.0005 and max. iterations = 300.

```python
all_theta_0, all_theta_1, all_losses, all_hx, optimum_theta = compute_gradient(
        X, y, learning_rate=0.0005, max_iterations=300
    )

print("All Theta_0 values:", all_theta_0)
print("All Theta_1 values:", all_theta_1)
print("All losses:", all_losses)
print("All hypothesis outputs:", all_hx)
print("Optimum Theta:", optimum_theta)
```

### Update your Batch GD for single variable  implementation (Batch variant) to be Momentum-Based GD and check your results

#### γ takes values between 0 and 1.

```python
def compute_momentum(momemtum_term,v_prev,gradient_vector,learning_rate):
    v_current = momemtum_term * v_prev + learning_rate * gradient_vector
    v_prev = v_current
    return v_current
```

```python
def update_momentum(theta,v):
    theta[0] -= v[0]
    theta[1] -= v[1]
    return theta
    
```

```python
def compute_momentum_gradient(X,y,learning_rate,momentum_term,max_iterations):
    loss =[] 
    theta = np.array([[0.0],[0.0]])
    v = np.array([[0.0],[0.0]])
    all_theta_0 = []  
    all_theta_1 = []  
    all_hx = []  
    tolerance = 0.001
    first_iteration = True
    for iteration in range(max_iterations):
        print(f"******************* Iteration: {iteration} *******************************")
        
        hx = predict(theta,X)
        all_hx.append(hx)
        print(f"h(x):{hx} \n")
        
        e = compute_error(hx,y)
        print(f"Error vector: {e} \n")
        
        j = compute_cost(e,X)
        loss.append(j)
        print(f"j = {j} \n")
        
        gradient_vector = compute_gradient_vector(e,X)
        print(f"Gradient_vector:  {gradient_vector} \n")
        
        gradient_vector_norm = compute_gradient_norm(gradient_vector)
        print(f"Gradient_vector_norm: {gradient_vector_norm} \n")

        v = compute_momentum(momentum_term,v,gradient_vector,learning_rate)
        print(f"v_0 : {v[0][0]}")
        print(f"v_1 : {v[1][0]}")
        
        all_theta_0.append(theta[0][0])
        all_theta_1.append(theta[1][0])
        
        if not first_iteration:
            if abs(loss[iteration] - loss[iteration - 1]) < tolerance:
                print(f"Converged at iteration {iteration} with cost: {loss[iteration]} \n")
                break
            if gradient_vector_norm < tolerance:
                print(f"Stopping at iteration {iteration} due to small gradient norm: {gradient_vector_norm} \n")
                break
        
        theta = update_momentum(theta,v)
        print(f"new theta_0 : {theta[0][0]}")
        print(f"new theta_1 : {theta[1][0]}")
        
        first_iteration = False
        
    print ("************************** Training Report ****************************")
    print(f"Gradient Descent converged after {len(all_theta_1)} epochs \n")
    print(f"Final theta_0: {theta[0][0]}")
    print(f"Final theta_1: {theta[1][0]} \n")

    return all_theta_0, all_theta_1, loss, all_hx, theta
        
```

### Update your Batch GD for one variable  implementation to be NAG and check your results

##### Very Important:
- <b><font color='red'> In order to calculate the gradient at theta_temp, you will need to calculate the prediction (hypothesis output) using theta_temp. </font></b> 
##### h_temp = theta0_temp + theta1_temp * X 

```python
def compute_NAG(X,y,learning_rate,momentum_term,max_iterations):
    loss =[] 
    theta = np.array([[0.0],[0.0]])
    theta_temp = np.array([[0.0],[0.0]])
    v = np.array([[0.0],[0.0]])
    all_theta_0 = []  
    all_theta_1 = []  
    all_hx = []  
    tolerance = 0.001
    first_iteration = True
    for iteration in range(max_iterations):
        print(f"******************* Iteration: {iteration} *******************************")

        hx = predict(theta,X)
        all_hx.append(hx)
        print(f"h(x):{hx} \n")
        
        e = compute_error(hx,y)
        print(f"Error vector: {e} \n")
        
        j = compute_cost(e,X)
        loss.append(j)
        print(f"j = {j} \n")
        
        gradient_vector = compute_gradient_vector(e,X)
        print(f"Gradient_vector:  {gradient_vector} \n")
        
        gradient_vector_norm = compute_gradient_norm(gradient_vector)
        print(f"Gradient_vector_norm: {gradient_vector_norm} \n")

        theta_temp = theta - momentum_term * v
        hx_temp = predict(theta_temp,X)
        e_temp = compute_error(hx_temp,y)
        gradient_vector_temp = compute_gradient_vector(e_temp,X)

        v = compute_momentum(momentum_term,v,gradient_vector_temp,learning_rate)
        print(f"v_0 : {v[0][0]}")
        print(f"v_1 : {v[1][0]}")
        
        all_theta_0.append(theta[0][0])
        all_theta_1.append(theta[1][0])
        
        if not first_iteration:
            if abs(loss[iteration] - loss[iteration - 1]) < tolerance:
                print(f"Converged at iteration {iteration} with cost: {loss[iteration]} \n")
                break
            if gradient_vector_norm < tolerance:
                print(f"Stopping at iteration {iteration} due to small gradient norm: {gradient_vector_norm} \n")
                break
        
        theta = update_momentum(theta,v)
        print(f"new theta_0 : {theta[0][0]}")
        print(f"new theta_1 : {theta[1][0]}")
        
        first_iteration = False
        
    print ("************************** Training Report ****************************")
    print(f"Gradient Descent converged after {len(all_theta_1)} epochs \n")
    print(f"Final theta_0: {theta[0][0]}")
    print(f"Final theta_1: {theta[1][0]} \n")

    return all_theta_0, all_theta_1, loss, all_hx, theta
        
```

```python
all_theta_0, all_theta_1, all_losses, all_hx, optimum_theta = compute_NAG(
        X, y, learning_rate=0.0005,momentum_term=0.9, max_iterations=300
    )

print("All Theta_0 values:", all_theta_0)
print("All Theta_1 values:", all_theta_1)
print("All losses:", all_losses)
print("All hypothesis outputs:", all_hx)
print("Optimum Theta:", optimum_theta)
```
